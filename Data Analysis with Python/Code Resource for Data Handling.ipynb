## Handling Data

```python
import pandas as pd
```

```python
# creating a series by passing a list
items = pd.Series(['chairs', 'labels', 'tables'])

items
```

```python
# creating a series by passing a list with the index
items = pd.Series(['chairs', 'labels', 'tables'], index =['OR 001', 'OR 002', 'OR 003'])

items
```

```python
# creating series
df = pd.Series(['Joe', 'Tom', 'Jane'])

df
```

### DataFrame

```python
dframe = {'Name':['Joe', 'Tom', 'Jane'],
          'Itmes':['Chairs', 'labels', ' tables']}

purchases = pd.DataFrame(dframe)

purchases
```
